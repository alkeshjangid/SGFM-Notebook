<h1>DEM manipulation and GIS workflow</h1>

<p>The objective is to capture and transform <b>DEM</b> data into a <b>CSV</b> file that will later be use to set a bunch of properties for a LECODE simulation. <br>
The data manipulation preserves the georeferencing information of the DEM, allowing later projection of the model in a GIS client (or google earth).
<br><br>
<u>Workflow description:</u>
<ol>
<li><a href="#DEM-provider">Download a DEM from the web (grid in lat/long coordinates)</a>, 
<li><a href="#ProjTest">Test the scene projection with GDAL</a>,  
<li><a href="#ReProj">Re-project the subset to UTM with GDAL in order to have a GeoTIF with pixels location in m</a>, 
<li><a href="#Crop">Crop the new projection and export the DEM at a selected resolution in a xyz format (in m)</a>,
<li><a href="#Transf">Perform the transformation to LECODE input format</a>,
<li><a href="#Ana">Analyse DEM for better definition of LECODE XmL input parameters</a>.
</ol>
</p>

<h2><a name="ProjTest" class="anchor" href="#ProjTest">
<span class="octicon octicon-link"></span></a>2. Test the scene projection with GDAL</h2>

<p>
First get information of the uploaded file and read general information about the upoaded file using <i>gdalinfo</i>. It will report all of the following (if known):
<ul>
<li>The format driver used to access the file.
<li>Raster size (in pixels and lines).
<li>The coordinate system for the file (in OGC WKT).
<li>The geotransform associated with the file (rotational coefficients are currently not reported).
<li>Corner coordinates in georeferenced, and if possible lat/long based on the full geotransform (but not GCPs).
<li>Ground control points.
<li>File wide (including subdatasets) metadata.
<li>Band data types.
<li>Band color interpretations.
<li>Band block size.
<li>Band descriptions.
<li>Band min/max values (internally known and possibly computed).
<li>Band checksum (if computation asked).
<li>Band NODATA value.
<li>Band overview resolutions available.
<li>Band unit type (i.e.. "meters" or "feet" for elevation bands).
<li>Band pseudo-color tables.
</ul>
<br>
<i><u>Note:</u></i><br>
If the projection information is only required use the <i>GetProjectionRef()</i> function. This function which returns the projection reference as a <a href="http://en.wikipedia.org/wiki/Well-known_text">WKT string</a>. It shows the image geodetic system (here <b>WGS 84</b>) and other projection information such as as well as the unit (degree) and ownership.
</p>

```python
"""
Open the file with GDAL.  What projection is it in?
"""
import os
import sys
from osgeo import gdal
from gdalconst import *
import matplotlib.pyplot as plt

# GDAL does not use python exceptions by default
gdal.UseExceptions()

# Create a Geotiff in case we started with an ArcInfo ASCII file
!gdalwarp -t_srs WGS84 ../SGFM-data/srtm_63_18/srtm_63_18.asc ../SGFM-data/srtm_63_18/srtm_63_18.tif

gtif = gdal.Open('../SGFM-data/srtm_63_18/srtm_63_18.tif',GA_ReadOnly)
    
# For a full overview of image information use gdalinfo command
!gdalinfo ../SGFM-data/srtm_63_18/srtm_63_18.tif

# If information on Projection Reference is only required used the following command
#gasc.GetProjectionRef()
```

Next we read the image data into a numpy array and plot it with matplotlib.

```python
arr = gtif.ReadAsArray()
trans = gtif.GetGeoTransform()
extent = (trans[0], trans[0] + gtif.RasterXSize*trans[1],
          trans[3] + gtif.RasterYSize*trans[5], trans[3])
figure(figsize = (5,5))
plt.imshow(arr, extent=extent)
plt.show()
```

<h2><a name="ReProj" class="anchor" href="#ReProj">
<span class="octicon octicon-link"></span></a>3. Re-project the subset to UTM</h2>

<p>
Now we re-project the image to UTM coordinate system in order to have the pixel location in metres.
<br>
To convert to Universal Transverse Mercator (UTM), <i>gdalwarp</i> requires the UTM zone of your image. You can use  <a href="http://epsg.io">EPSG</a> webinterface to check if your UTM zone is correct. Then use <i>gdalinfo</i> to check any issues in the re-projected UTM image.
</p>

```python
!gdalwarp -t_srs '+proj=utm +zone=52 + south +datum=WGS84' ../SGFM-data/srtm_63_18/srtm_63_18.tif ../SGFM-data/srtm_63_18/utm_63_18.tif

# For a full overview of image information use gdalinfo command
!gdalinfo ../SGFM-data/srtm_63_18/utm_63_18.tif
```

Have a look at the projected image.
<br>

```python
gutm = gdal.Open('../SGFM-data/srtm_63_18/utm_63_18.tif',GA_ReadOnly)
arr_utm = gutm.ReadAsArray()
trans_utm = gutm.GetGeoTransform()
uextent = (trans_utm[0], trans_utm[0] + gutm.RasterXSize*trans_utm[1],
          trans_utm[3] + gutm.RasterYSize*trans_utm[5], trans_utm[3])
figure(figsize = (5,5))
plt.imshow(arr_utm, extent=uextent)
plt.show()
```

<h2><a name="Crop" class="anchor" href="#Crop">
<span class="octicon octicon-link"></span></a>4. Clip the region to simulation area</h2>

<p>
Now that the DEM has been projected in UTM, we will crop it to fit with the extent of the region that we want to simulate with LECODE. <br>
In this notebook we want to defined a region of <i>80 km</i> large and <i>100 km</i> long around the following coordinate <b>(132°,-28°)</b> using a resolution of <i>200 m</i>.
<br>
The long/lat coordinate is projected in the correct coordinates system using <b>PyProj</b> package. Then we display it on the map with a marker. 
</p>

```python
from pyproj import Proj
#p=Proj(proj='utm',zone=10,ellps='WGS84')
p = Proj(proj='utm',zone='52S',ellps='WGS84')
p.srs
```

```python
# Point coordinate projected in UTM
ox,oy = p(132,-28)
print 'UTM centre coordinate: ',ox,oy

# Simulation lenght, width
lg=80000
wd=100000

# Define simulation xmin, xmax
xmin = ox-lg/2
xmax = ox+lg/2
# Define simulation ymin, ymax
ymin = oy-wd/2
ymax = oy+wd/2

cliped_file = '../SGFM-data/srtm_63_18/clip_utm.tif'
os.remove(cliped_file) if os.path.exists(cliped_file) else None

print 'Simulation box extent: ',int(xmin),int(ymin),int(xmax),int(ymax)
```

We will use the extent of the box to populate the <i>gdalwarp</i> function and to define the resolution of the image (set to 200m).
<br>
<b><u>Attention</u>: you will have to change the <i>gdalwarp</i> options in the script!</b>

```python
# Crop the region at the given resolution (by defining new resolution in the tr option)
!gdalwarp -te 755043 -3150830 835043 -3050830 -tr 200 200 -tap -r bilinear ../SGFM-data/srtm_63_18/utm_63_18.tif ../SGFM-data/srtm_63_18/clip_utm.tif

# Open the file and check projection information
gclip = gdal.Open(cliped_file,GA_ReadOnly)
if gclip is None:
    print 'Could not open image'
    sys.exit(1)
!gdalinfo -mm ../SGFM-data/srtm_63_18/clip_utm.tif

# Display the figure with the simulation area
arr_clip = gclip.ReadAsArray()
trans_clip = gclip.GetGeoTransform()
cextent = (trans_clip[0], trans_clip[0] + gclip.RasterXSize*trans_clip[1],
          trans_clip[3] + gclip.RasterYSize*trans_clip[5], trans_clip[3])
figure(figsize = (7,7))
plt.imshow(arr_clip, extent=cextent)
plt.show()
```

Use a predefined text color scale file (some predefined color scale are proposed in the <code>color-scale</code> foler) to produce a PNG file of the clipped region.

<h2><a name="Transf" class="anchor" href="#Transf">
<span class="octicon octicon-link"></span></a>5. Perform the transformation to LECODE input format</h2>

<p>
In this step we will generate the topographic and initial deposit files for LECODE simulation based on the chosen area.
<br>
First we crate and ASCII grid that will serve durig the export process. Then the <code>lecode_basic_input</code> function will create the required LECODE files.
<pre><code>
lecode_basic_input name_ascii_grid thickness_init_layer nb_material_type init_layer_hardness
</code></pre>
The function requires 3 parameters and an optional one:
<ul>
<li> <code>name_ascii_grid</code> name of the ASCII grid generated using <code>gdal_translate</code>
<li> <code>thickness_init_layer</code> thickness of the initial sedimentary layer in metres,
<li> <code>nb_material_type</code> number of sediment that will be defined in LECODE XmL input file,
<li> <code>init_layer_hardness</code> is optional (default is 1) and enable the definition of a hardness value for basement sedimentary layer.
</ul>

</p>

```python
import os, sys

# Create an ASCII grid from the simulation area
!gdal_translate -of AAIGrid ../SGFM-data/srtm_63_18/clip_utm.tif ../SGFM-data/srtm_63_18/dem.asc 

# Perform the transformation to LECODE input format
os.system('lecode_basic_input ../SGFM-data/srtm_63_18/dem.asc 50 5 10')
print "LECODE input have been created, check for .nodes & .dep files"
```

<h2><a name="Ana" class="anchor" href="#Ana">
<span class="octicon octicon-link"></span></a>6. Analyse DEM for better definition of LECODE XmL input parameters</h2>

<h3>DEM relief</h3>
<p>
To have a look at the elevation range use gdalinfo command and look within the answer for <code>Computed Min/Max</code>.
<br>
Base on the range you can define a text color scale file (some predefined color scales are proposed in the <code>color-scale</code> foler) to produce a PNG file of the clipped region.  
</p>

```python
# Find information about min/max elevation 
!gdalinfo -mm ../SGFM-data/srtm_63_18/clip_utm.tif

# Based on elevation extent use an appropriate color scale to create the DEM map
!gdaldem color-relief ../SGFM-data/srtm_63_18/clip_utm.tif ../SGFM-data/color-scale/color_ramp.txt ../SGFM-data/srtm_63_18/region.png -of PNG 
```

<h3>DEM slope</h3>
<p>
To produce the DEM slope we use <i>gdaldem</i> associated with <i>slope</i> command line and the <i>-p</i> option as well which specifies that the calculated slope will be expressed as percent slope.
<br>
To check the slope range you can use <i>gdalinfo</i> command and look within the answer for <code>Computed Min/Max</code>.
<br>
Base on the range you can define a text color scale file (some predefined color scales are proposed in the <code>color-scale</code> foler) to produce a PNG file of the clipped region.  
</p>

```python
# From a DEM raster, output a 32-bit float raster with slope values (specifies in percent slope)
!gdaldem slope ../SGFM-data/srtm_63_18/clip_utm.tif ../SGFM-data/srtm_63_18/slope_utm.tif -p 

# Find information about min/max slope  
!gdalinfo -mm ../SGFM-data/srtm_63_18/slope_utm.tif

# Based on elevation extent use an appropriate color scale to create the DEM slope map
!gdaldem color-relief ../SGFM-data/srtm_63_18/slope_utm.tif ../SGFM-data/color-scale/color_slope2.txt ../SGFM-data/srtm_63_18/slopeshade.png -of PNG
```

```python
from IPython.core.display import Image 

# Show the relief map
Image(filename='../SGFM-data/srtm_63_18/region.png')
```

```python
# Show the slope map
Image(filename='../SGFM-data/srtm_63_18/slopeshade.png')
```

<h3>Flow accumulation map</h3>
<p>
LECODE uses flow accumulation maps to defined several parameters such as the adaptive TIN resolution or to trigger mass movements.  
<br>
The next step will create the initial flow accumulation map that could be use to define these parameters in the XmL input file.
</p>

```python
import os, sys

# Perform Planchon fill depression algorithm
os.system('fill_depression ../SGFM-data/srtm_63_18/dem.asc ../SGFM-data/srtm_63_18/fill.asc 0.001')

# Run the multiple flow direction algorithm
os.system('flow_acc ../SGFM-data/srtm_63_18/fill.asc ../SGFM-data/srtm_63_18/facc.asc')

# Transform the MFD flow accumulation map in Geotiff
!gdal_translate -of "GTiff" ../SGFM-data/srtm_63_18/facc.asc ../SGFM-data/srtm_63_18/facc.tif
```

```python
# Open the file
gflow = gdal.Open('../SGFM-data/srtm_63_18/facc.tif',GA_ReadOnly)

# Get the flow accumulation range
!gdalinfo -mm ../SGFM-data/srtm_63_18/facc.tif
```

```python
# Based on flow accumulation extent use an appropriate color scale to create the DEM flow accumulation map
!gdaldem color-relief ../SGFM-data/srtm_63_18/facc.tif ../SGFM-data/color-scale/color_facc2.txt ../SGFM-data/srtm_63_18/facc.png -of PNG
```
